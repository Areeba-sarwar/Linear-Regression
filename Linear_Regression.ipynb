```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
dataset= pd.read_csv(r"C:\Users\Ubaid\Downloads\archive (16)\Salary_dataset.csv")
```

```python
# Compute correlations
correlation = dataset.corr(numeric_only=True)
# Correlation with target 
print(correlation["Salary"].sort_values(ascending=False))
```

```python
X = dataset["YearsExperience"].values.reshape(-1, 1)
Y= dataset["Salary"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X, Y)
```

```python
y_pred = model.predict(X_test)
```

```python
#Evaluation
print("R² Score:", r2_score(y_test, y_pred))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted')
plt.title('Linear Regression: YearsExperience vs. Salary')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.legend()
plt.grid(True)
plt.show()
```

```python
user_input = float(input("Enter Years of Experience to predict Salary: "))
input_array = np.array(user_input).reshape(-1, 1)
predicted_salary = model.predict(input_array)
print(f"Predicted Salary for {user_input} years of experience is: {predicted_salary[0]:.2f}")
```
